```python
# object(categorical columns)
categorical_features = df.select_dtypes(include =['object']).axes[1]
```

```python
# one hot encoding and also concating the rest of columns.
# dropping the first columns which is usually considered for better results

for col in categorical_features:
    df = pd.concat([df , pd.get_dummies(df[col], prefix = col , prefix_sep = ':', drop_first = True)] , axis = 1)
    df.drop(col , axis = 1 , inplace = True)
```

```python
df.head()
```

```python
df.info()
```

```python
X = df.drop(' income' , axis = 1).copy()
y = df[' income']
```

```python
X.head()
```

```python
y.value_counts()
```

## **DATA SPLITING**

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, y, test_size=0.2)
```

## **MODEL TRAINING**

### **1.  Decision Tree**

```python
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier()
dtc.fit(X_train, Y_train)
```

```python
Y_pred = dtc.predict(X_test)
c1= dtc.score(X_test, Y_test)
```

### **2.  Logistic Regression**

```python
from sklearn.linear_model import LogisticRegression
lg = LogisticRegression(max_iter=10000)
lg.fit(X_train, Y_train)
```

```python
Y_pred = lg.predict(X_test)
c2= lg.score(X_test, Y_test)
```

### **3. SVM**

```python
from sklearn import svm
sv = svm.SVC()
sv.fit(X_train, Y_train)
```

```python
Y_pred = lg.predict(X_test)
c3= lg.score(X_test, Y_test)
```

### **4. Gaussian Naive Bayes**

```python
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
gnb.fit(X_train, Y_train)
```

```python
Y_pred = gnb.predict(X_test)
c4= gnb.score(X_test, Y_test)
```

### **5. Random Forest Classification**

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier()
rfc.fit(X_train, Y_train)
```

```python
Y_pred = rfc.predict(X_test)
c5= rfc.score(X_test, Y_test)
```

```python
name = (' Decision Tree', ' Logistic Regression',' SVM',' Gaussian Naive Bayes',' Random Forest Classification')
score = (c1,c2,c3,c4,c5)
```

```python
dfc = pd.DataFrame()
dfc['Name']= name
dfc['Score']= score
dfc
```

```python
sns.barplot(x = "Name", y = 'Score', data = dfc, palette = 'viridis')
plt.xticks(rotation=90)
plt.show()
```

## **CONCLUSION**


From the above graph and table we can conclude that **Random Forest Classification** is the best model for this dataset for regression which has the score of about 85%
